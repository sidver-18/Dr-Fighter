```python
import pygame
import random
import math
import time

pygame.init()
window=pygame.display.set_mode((600,600))
pygame.display.set_caption("Dr. Fighter")

background = pygame.image.load("DrImg/background.png")
Dright=[pygame.image.load('DrImg/R1.png'), pygame.image.load('DrImg/R2.png'), pygame.image.load('DrImg/R3.png'), pygame.image.load('DrImg/R4.png'), pygame.image.load('DrImg/R5.png'), pygame.image.load('DrImg/R6.png'), pygame.image.load('DrImg/R7.png'), pygame.image.load('DrImg/R8.png'), pygame.image.load('DrImg/R9.png')]
Dleft=[pygame.image.load('DrImg/L1.png'), pygame.image.load('DrImg/L2.png'), pygame.image.load('DrImg/L3.png'), pygame.image.load('DrImg/L4.png'), pygame.image.load('DrImg/L5.png'), pygame.image.load('DrImg/L6.png'), pygame.image.load('DrImg/L7.png'), pygame.image.load('DrImg/L8.png'), pygame.image.load('DrImg/L9.png')]

bubbleSound = pygame.mixer.Sound('DrImg/bullet.wav')

clock=pygame.time.Clock()

Dsteps=0
x = 5
y = 520
width = 64
height = 64
vel = 5
last=0
scoreValue=0
text="WE SAVED"

virusImg = pygame.image.load('DrImg/virus.png')
no_of_virus = 15
screenwidth = 600
virus_x = []
virus_y = []
virus_vel = []


for i in range(no_of_virus):
    virus_x.append(random.randint(0,560))
    virus_y.append(random.randint(10,200))
    virus_vel.append(10)

                  
                    
font = pygame.font.Font("freesansbold.ttf", 20)
endfont=pygame.font.Font("freesansbold.ttf", 40)
show = pygame.font.Font("freesansbold.ttf",20)

def show_score(x, y):
    global scoreValue
    score = font.render("Score :" + str(scoreValue), True, (255, 255, 255))
    window.blit(score, (x, y))
    
def gameOver():
    global text
    end = endfont.render(text, True, (255, 255, 255))
    window.blit(end, (150, 250))
    
def show_text(x,y,color):
    global text
    main=font.render(text,True,color)
    window.blit(main,(x,y))
    
        
left = False
right = False    
play=True
brun=False

bubbleImg = pygame.image.load('DrImg/bubble.png')
bubbleX = 0
bubbleY = 480
bubbleX_change = 0
bubbleY_change = 30

def Dmovement():
    global Dsteps
    global last
    if Dsteps+1>=27:
        Dsteps=0
    if left:
        window.blit(Dleft[Dsteps//3],(x,y))
        last=1
        Dsteps+=1
    elif right:
        window.blit(Dright[Dsteps//3],(x,y))
        Dsteps+=1
        last=0
    else:
        if(last==0):
            window.blit(Dright[0],(x,y))
            Dsteps==0
        elif (last==1):
            window.blit(Dleft[0],(x,y))
            Dsteps==0
def Bubble(x,y):
    window.blit(bubbleImg,(x+30,y+10))
    

t_end = time.time() + 40
t_win=t_end+2
    
while time.time() <t_end:
    clock.tick(27)
    window.fill((0,0,0))
    window.blit(background,(0,0))
    text="SAFE ZONE"
    show_text(5,480,(0,50,255))
    text="DANGER ZONE"
    show_text(5,10,(255,0,0))
    
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            t_end=-1
        
    keys=pygame.key.get_pressed()
    if keys[pygame.K_SPACE]:
        brun=True
        bubbleSound.play()
         
    if brun: 
        bubbleX=x
        Bubble(bubbleX,bubbleY)
        bubbleY-=bubbleY_change
    if bubbleY<=0:
        brun=False
        bubbleY=450
            
    
    if keys[pygame.K_LEFT] and x > vel:
        x -= vel
        left=True
        right=False
    elif keys[pygame.K_RIGHT] and x < 600 - width - vel:
        x += vel
        left=False
        right=True
    else:
        left=False
        right=False
        
    Dmovement()
    for i in range(no_of_virus):
        virus_x[i] += virus_vel[i]
        if virus_x[i] <= 0:
            virus_vel[i] *= -1
        if virus_x[i] >= screenwidth-40:    
            virus_vel[i] *= -1
            virus_y[i] += 30
    #collision     
    for i in range(no_of_virus):
        Distance = math.sqrt(math.pow((bubbleX-virus_x[i]),2) + math.pow((bubbleY-virus_y[i]),2))
        if Distance < 27 and virus_y[i]<300:
            scoreValue+=1
            virus_x[i] = random.randint(0,560)
            virus_y[i] = random.randint(10,200)
    
    #loosing condition
    for i in range(no_of_virus):
        
        if virus_y[i] > 500:
            for i in range(no_of_virus):
                virus_y[i] = 2000
                text="DOCTOR DIED"
                gameOver()
                t_end=-1
                break
    
    for i in range(no_of_virus):
        window.blit(virusImg,(virus_x[i],virus_y[i]))
    
    show_score(500,10)
    pygame.display.update()

while(time.time()<t_win and not(t_end==-1)):
    text="WE SAVED"
    gameOver()
    pygame.display.update()

pygame.quit()
```
